```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('/Users/bennyogayonne/Downloads/waste.csv')
```

```python
print(df)
```

```python
##THE MOST COMMON WASTE SOURCE
most_common_source = df['Main_Sources'].mode()[0]
print(f'The most common source of waste is: {most_common_source}')
```

```python
##THE COUNTRIES WITH THE LEAST / MOST WASTE
```

```python
most_plastics = df.groupby('Country')['Total_Plastic_Waste_MT'].sum()
```

```python
max_plastic_country = most_plastics.idxmax()
max_plastic_number = most_plastics.max()
```

```python
print(f'The country with the most plastic waste is {max_plastic_country} with {max_plastic_number} million tons of plastic.')
```

```python
least_plastics = df.groupby('Country')['Total_Plastic_Waste_MT'].sum()
```

```python
least_plastic_country = least_plastics.idxmin()
least_plastic_number = least_plastics.min()
```

```python
print(f'The country with the least plastic waste is {least_plastic_country} with {least_plastic_number} million tons.')
```

```python
grouped_data = df.groupby('Country')['Total_Plastic_Waste_MT'].sum().reset_index()
top_10 = grouped_data.sort_values(by='Total_Plastic_Waste_MT', ascending=False).head(10)
plt.figure(figsize=(12, 6))  
plt.bar(top_10['Country'], top_10['Total_Plastic_Waste_MT'], color='red')  
plt.xlabel('Country')  
plt.ylabel('Total Plastic Waste (in tons)') 
plt.title('Top 10 Countries by Total Plastic Waste')  
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()  
```

```python
##COUNTRIES RECYCLING RATES
```

```python
average_recycling = df['Recycling_Rate'].mean()
```

```python
print(f'The average recycling rate is {average_recycling :.2f} worldwide')
```

```python
most_recycling = df.groupby('Country')['Recycling_Rate'].sum()
```

```python
most_recycling_country = most_recycling.idxmax()
most_recycling_number = most_recycling.max()
```

```python
print(f'The country with the highest recycyling rate is {most_recycling_country} with a recycling rate of {most_recycling_number}.') 
```

```python
least_recycling = df.groupby('Country')['Recycling_Rate'].sum()
```

```python
least_recycling_country = least_recycling.idxmin()
least_recycling_number = least_recycling.min()
```

```python
print(f'The country with the lowest recycling rate is {least_recycling_country} with a rate of {least_recycling_number}.')
```

```python
grouped_data = df.groupby('Country')['Recycling_Rate'].sum().reset_index()
top_10 = grouped_data.sort_values(by='Recycling_Rate', ascending=False).head(10)
plt.figure(figsize=(12, 6))  
plt.bar(top_10['Country'], top_10['Recycling_Rate'], color='green')  
plt.xlabel('Country')  
plt.ylabel('Recycle Rate') 
plt.title('Top 10 Countries by Recycle Rate')  
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()  
```
